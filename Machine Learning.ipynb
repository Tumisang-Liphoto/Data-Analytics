# Machine Learning

# Agenda

- **Machine Learning** is an application of artificial intelligence (AI) that provides systems the ability to automatically learn and improve from experience without being explicitly programmed. Machine learning focuses on the development of computer programs that can access data and use it learn for themselves. The process of learning begins with observations or data, such as examples, direct experience, or instruction, in order to look for patterns in data and make better decisions in the future based on the examples that we provide. The primary aim is to allow the computers learn automatically without human intervention or assistance and adjust actions accordingly.

<font color = blue> In a simplified manner, ML is the development of a mapping $f$, that interpretes the relationships (if any exists) between a set of inputs $x$ and a result $Y$. This is such that the result of the learning process is $Y=f(x)$; the machine is said to have learned when it is able to relate unseen input to a given output to a given degree of error.</font>

- **Supervised Machine Learning** algorithms can apply what has been learned in the past to new data using **labeled** examples to predict future events. Starting from the analysis of a known **training dataset**, the learning algorithm produces an inferred function to make predictions about the output values. The system is able to provide targets for any new input after sufficient training. The learning algorithm can also compare its output with the correct, intended output and find errors in order to modify the model accordingly.

- In contrast, **Unsupervised Machine Learning** algorithms are used when the information used to train is neither classified nor labeled. Unsupervised learning studies how systems can infer a function to describe a hidden structure from unlabeled data. The system doesnt figure out the right output, but it explores the data and can draw inferences from datasets to describe hidden structures from unlabeled data.

<h3> Problems of training testing on the same data </h3>
 

1. Goal is to estimate on out of sample data
1. Maximising on acuracy on same data results in over complex model
1. Over complex model overfit on the training data



- The model has **memorised** the data as opposed to **learning** from it

<h1> Instead of training and testing on same data, we implement <font color = red> TRAIN/TEST SPLIT </font> <h1>

1. Split the data set into two pieces: **training set** and **testing set**
1. Train the model on the **training set**
1. Test and evaluate the model on the **testing set**

# Step 1: Split the data set into training and testing datasets

```python
#import train/test split library

from sklearn.model_selection import train_test_split

#we are to use 40% of the data as the test data
#The data to make up the 40% is random
#for simplisity we set a paramenter "random state" to ensure the same random data is always selected 
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.4)
```

```python
#print the shapes of training and testin data sets
print(X_train.shape)
print(X_test.shape)
```

```python
print(y_train.shape)
print(y_test.shape)
```

#  Step 2: Train the model on the training set

```python
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

#  Step 3: Make prediction on the testing data

```python
y_pred = logreg.predict(X_test)
#compare the actual resposes (y_test) with predicted responses (y_pred)
print (metrics.accuracy_score(y_test,y_pred))
```

<h3> Repeat for KNN with K=1 <h3>

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print (metrics.accuracy_score(y_test,y_pred))
```

<h3> Repeat for KNN with K=5 <h3>

```python
knn_5 = KNeighborsClassifier(n_neighbors=5)
knn_5.fit(X_train, y_train)
y_pred = knn_5.predict(X_test)
print (metrics.accuracy_score(y_test,y_pred))
```

```python
knn_15 = KNeighborsClassifier(n_neighbors=15)
knn_15.fit(X_train, y_train)
y_pred = knn_15.predict(X_test)
print (metrics.accuracy_score(y_test,y_pred))
```

```python
knn_75 = KNeighborsClassifier(n_neighbors=75)
knn_75.fit(X_train, y_train)
y_pred = knn_75.predict(X_test)
print (metrics.accuracy_score(y_test,y_pred))
```

<h3> Can we find the best value of K? <h3>

```python
#try k = 1 to k = 25 and record the corresponding testing accuracy

k_range = range(1,26)
scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k) 
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    scores.append(metrics.accuracy_score(y_test,y_pred))
    
print(scores)
```

```python
#import library to "see" the results

import matplotlib.pyplot as plt

#allow graph to appear inline
%matplotlib inline

#plot the relationship between k and testing accuracy
plt.plot(k_range,scores)
plt.xlabel("Value of k for KNN")
plt.ylabel("Testing Accuracy")
```
